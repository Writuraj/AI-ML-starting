```python
def train_linear_regression(X,y):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones,X])

    XTX = X.T.dot(X)
    XTXinv = np.linalg.inv(XTX)
    W_make = XTXinv.dot(X.T).dot(y)
    return W_make[1:],W_make[0]
```

```python
train_linear_regression(X,y)
```
