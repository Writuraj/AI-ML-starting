```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
```

```python
y=2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xlim(0,6)
plt.ylim(0,15)
plt.title('string title')
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
```

```python
a = np.linspace(0,10,11)
b=a**4
```

```python
fig=plt.figure()
#Large axes
axes1=fig.add_axes([0,0,1,1])
#Small axes 
axes2=fig.add_axes([0.3,0.3,0.25,0.25])
axes1.plot(a,b)
axes2.plot(x,y)
```

## Subplot

```python
fig,axes=plt.subplots(nrows=2,ncols=2)

axes[0,0].plot(x,y)
 
axes[0,1].plot(a,b)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1)
axes[0].plot(x,y)
axes[1].plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0,0].plot(x,y)
axes[0,1].plot(x,y)
axes[1,0].plot(x,y)
axes[1,1].plot(x,y)
fig.subplots_adjust(wspace=0.9,hspace=1)

fig.suptitle("My subplots")
```

## Styling

```python
x=np.linspace(0,11,10)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,label= "X vs X")
ax.plot(x,x**2,color='red',label= "X vs X^2",lw=4,ls=':')
ax.legend()
```

## self designing line

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
#ax.plot(x,x,label= "X vs X")
ax.plot(x,x,color='red',label= "X vs X^2",lw=4,ls=':',marker='o')
ax.plot(x,x+1,color='red',label= "X vs X^2",lw=4,marker='o',ls='--',ms=15,markerfacecolor='orange',markeredgewidth=4,markeredgecolor='yellow')
```
