# Dot Product

```python
def vector_vector_multiplication(a,b):
    assert a.shape[0]==b.shape[0]
    n=a.shape[0]
    result=0.0
    for i in range (n):
        result = result + a[i] * b[i]
    return result
```

```python
vector_vector_multiplication(a,b)
```
